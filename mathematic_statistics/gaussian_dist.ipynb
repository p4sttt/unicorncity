Нормальное распредление называние распредление случайной величины $x \sim \mathcal{N(\mu, \sigma)}$ с плотность распредение

$$
f_{x}(t)=\frac{1}{\sqrt{ 2\pi\sigma^2 }}\exp \left( -\frac{(t-\mu)^2}{2\sigma^2} \right) 
$$

Соотвественно

$$
\mathbb{E}\left[ X \right] =\int_{-\infty}^{\infty} tf_{x}(t)dt=\mu
$$

и

$$
\operatorname{Var}\left[ X \right] =\sigma^2
$$

(TODO: график нормального распредления через python)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

def normal_pdf(x, mean=0.0, std=1.0):
    return (1 / (std * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std)**2)

def plot_normal(ax, mean=0.0, std=1.0, x_range=(-4, 4), n=1000, label=None):
    x = np.linspace(x_range[0], x_range[1], n)
    y = normal_pdf(x, mean=mean, std=std)
    sns.lineplot(x=x, y=y, ax=ax, label=label if label is not None else f"μ={mean}, σ={std}")
    return ax

fig, ax = plt.subplots(figsize=(8, 5))
plot_normal(ax, mean=0, std=1, x_range=(-4, 4), label="Standard normal (μ=0, σ=1)")
plot_normal(ax, mean=1, std=0.5, x_range=(-4, 4), label="Shifted narrow (μ=1, σ=0.5)")
ax.set_title("Normal Distributions")
ax.set_xlabel("x")
ax.set_ylabel("f(x)")
ax.legend()
plt.show()
```
